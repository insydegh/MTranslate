```
i = 0 
b = '/content/gdrive/My Drive/ru_data/icemarch'
X = []
```

```
X
```

```

for file in soundlist:
    a = str(file)
    loaded_sound = []
    Y, sr = librosa.load(b+'/'+a)
    loaded_sound.append(librosa.feature.mfcc(Y, sr=sr))
    X = X + loaded_sound
    
```

```
X
```

```
type(X)
```

```
Y = []
for i in range(0,1916):
    Y.append(russian.loc[i, 'trans'])
```

```
Y = np.asarray(Y)
X = np.array(X, ndmin = -1)
```

```
Y
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```
model = Sequential()
model.add(LSTM(units=256,return_sequences=True))
model.add(LSTM(units=128,return_sequences=True))
model.add(LSTM(units=64,return_sequences=True))
model.add(LSTM(units=32))
model.add(Dense(units=128, activation='relu'))

model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(np.asarray(X_train), np.asarray(Y_train), epochs=10000,validation_data=(np.asarray(X_test), np.asarray(Y_test)))
```
